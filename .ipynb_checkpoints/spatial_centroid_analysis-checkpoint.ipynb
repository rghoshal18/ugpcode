```python
import pandas as pd
import numpy as np
```

```python
def centroids(df,value):
    
    ## df description
    ## 1st dimension year
    ## 2nd dimension lat
    ## 3rd dimension lon
    ## 4th value

    df["transformed"] = 1
    
    for lat in df["lat"].unique():
        for lon in df["lon"].unique():
            dfmax = df[value][ (df["lat"] == lat) & (df["lon"] == lon) ].max()
            dfmin = df[value][ (df["lat"] == lat) & (df["lon"] == lon) ].min()
            for year in df["year"].unique():
                if dfmax != dfmin:
                    df["transformed"][df["year"] == year] = (df[value][df["year"] == year] - dfmin)/(dfmax - dfmin)
    coordx=[]
    coordy=[]
    for year in df["year"].unique():
        coordx.append(np.dot(df["transformed"][df["year"] == year],df["lon"][df["year"] == year])/df["transformed"][df["year"] == year].sum())
        coordy.append(np.dot(df["transformed"][df["year"] == year],df["lat"][df["year"] == year])/df["transformed"][df["year"] == year].sum())
        
    return coordx,coordy
```

```python
data = pd.read_csv("/mnt/d/UGP/datasets/data.csv",index_col="Unnamed: 0").drop(columns=["sla","enso"])
```

```python
data["year"] = pd.DatetimeIndex(data["time"]).year
```

```python
centroids(data,"precip")
```
