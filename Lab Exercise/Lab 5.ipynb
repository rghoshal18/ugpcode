```python
k=np.loadtxt('sawww.txt',delimiter=";")
Q=pd.DataFrame(k,columns='Year-Month Mean-sea-level Days-missing Flag-attention'.split())

Y=k


print(Y[0][0])
Mean = 0
print(Y.shape)

for i in range(0, Y.shape[0]):
    temp = str(Y[i][0])
    temp.split()
    temp = temp[0]+temp[1]+temp[2]+temp[3]
    f = int(temp)
    #print(f)
    if(f>=1878 and f<=1936):
        Y[i][1] = Y[i][1] + 3850
    elif(f>1936):
        Y[i][1] = Y[i][1] + 4460
        Mean=Mean+Y[i][1]
        
    print(Y[i][1])   
    
print('Mean: ')
Z=mean/Y.shape[0]
print(Z)
```
